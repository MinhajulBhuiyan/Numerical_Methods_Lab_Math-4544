# **Lab 1: Bisection Method**<br>

The bisection method (also known as the binary search method) is a method of finding roots of a non-linear function. The assumption here is that the function is real continuous, otherwise the bisection method will not work. It is based on the **Intermediate Value Theorem** which states that if $f(x)$ is a real continuous function, and if $f(x_l)f(x_u)< 0$ then there must be a root between $x_l$ and $x_u$.

````python
import numpy as np
import matplotlib.pyplot as plt

# Function
def func(f, x):
    res = 0
    n = len(f)

    for i in range(n):
        coeff = f[i]
        power = n - i - 1
        value = coeff * (x ** power)
        res += value

    return res

# Bisection method
def bisection_method(f, xl, xu, epsilon, iterations):
    iter_count = 0
    xr = xl  # Initial midpoint
    errors = []
    significant_digits = []

    while iter_count < iterations:
        xr_old = xr
        xr = (xl + xu) / 2.0

        f_xl = func(f, xl)
        f_xr = func(f, xr)

        if f_xl * f_xr < 0:
            xu = xr  # The root is in the lower half
        else:
            xl = xr  # The root is in the upper half

        if xr != 0:
            error = abs((xr - xr_old) / xr) * 100
            errors.append(error)

            # Significant digits
            sig_digits = 2-np.log10(error / 0.5)
            significant_digits.append(sig_digits)
        else:
            error = float('inf')
            significant_digits.append(0)

        if error < epsilon:
            break

        iter_count += 1

    return xr, significant_digits

# Driver code
f = np.array([1, -2, 0, 4])  # Polynomial coefficients
xl = -2
xu = 3
epsilon = 0.005
MAX_ITERS = 100

xr, significant_digits = bisection_method(f, xl, xu, epsilon, MAX_ITERS)
print(f"Root ≈ {xr}")

# Plot the graph
plt.plot(range(1, len(significant_digits) + 1), significant_digits, marker='o')
plt.title("Number of Iterations vs Number of Significant Digits (m)")
plt.xlabel("Number of Iterations")
plt.ylabel("Number of Significant Digits (m)")
plt.grid(True)
plt.show()
````

Plot a graph with Number of Iterations along the $X$-axis and Root Approximation $(x_m)$ along the $Y$-axis.

````python
import numpy as np
import matplotlib.pyplot as plt

# Function
def func(f, x):
    res = 0
    n = len(f)

    for i in range(n):
        coeff = f[i]
        power = n - i - 1
        value = coeff * (x ** power)
        res += value

    return res

# Bisection method to find the root
def bisection_method(f, xl, xu, epsilon, iterations):
    iter_count = 0
    xr = xl  # Initial midpoint
    approximations = []

    while iter_count < iterations:
        xr_old = xr
        xr = (xl + xu) / 2.0

        # Evaluate function at xl and xr
        f_xl = func(f, xl)
        f_xr = func(f, xr)

        if f_xl * f_xr < 0:
            xu = xr  # The root is in the lower half
        else:
            xl = xr  # The root is in the upper half

        # Store the current approximation
        approximations.append(xr)

        if abs((xr - xr_old) / xr) * 100 < epsilon:
            break

        iter_count += 1

    return xr, approximations

# Driver code
f = np.array([1, -2, 0, 4])  # Polynomial coefficients
xl = -2
xu = 3
epsilon = 0.005
MAX_ITERS = 100

xr, approximations = bisection_method(f, xl, xu, epsilon, MAX_ITERS)
print(f"Root ≈ {xr}")

# Plot the graph
plt.plot(range(1, len(approximations) + 1), approximations, marker='o')
plt.title("Number of Iterations vs Root Approximation (xm)")
plt.xlabel("Number of Iterations")
plt.ylabel("Root Approximation (xm)")
plt.grid(True)
plt.show()
````
